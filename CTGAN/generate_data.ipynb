```python
import pickle
```

```python
with open("ctgan_model_small_model.pkl", 'rb') as f:
    model = pickle.load(f)
```

```python
synthetic_data = model.sample(2200)
```

```python
synthetic_data.to_csv("synthetic_data.csv", index=False)
```

```python
# Change PTID to catergorical and renumber it
synthetic_data['PTID'] = synthetic_data['PTID'].astype('category').cat.codes
synthetic_data.sort_values('PTID', inplace=True)
```

```python
synthetic_data.to_csv("synthetic_data.csv", index=False)
```
